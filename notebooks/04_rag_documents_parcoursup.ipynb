# Initialisation

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/parcoursup_2026_enriched.csv")
```

# Fonction pour l'ajout d'une colonne utile au RAG

```python
def build_rag_doc(row):
    parts = []

    # En-tête
    if pd.notna(row["name_formation"]) and str(row["name_formation"]).strip() != "":
        parts.append(f"Nom de la formation : {row['name_formation']}")
    if pd.notna(row["type_formation"]) and str(row["type_formation"]).strip() != "":
        parts.append(f"Type de formation : {row['type_formation']}")
    if pd.notna(row["name_etablissement"]) and str(row["name_etablissement"]).strip() != "":
        parts.append(
            f"Établissement : {row['name_etablissement']}, "
            f"{row.get('type_etablissement', '')} "
            f"({row.get('commune', '')}, {row.get('departement', '')}, {row.get('region', '')})"
        )
    if pd.notna(row["mentions_specialites"]) and str(row["mentions_specialites"]).strip() != "":
        parts.append(f"Mentions / spécialités : {row['mentions_specialites']}")

    # Texte scrappé
    if pd.notna(row["presentation"]) and str(row["presentation"]).strip() != "":
        parts.append(f"Présentation : {row['presentation']}")
    if pd.notna(row["criteres_entree"]) and str(row["criteres_entree"]).strip() != "":
        parts.append(f"Critères d’entrée : {row['criteres_entree']}")
    if pd.notna(row["poursuite_etudes"]) and str(row["poursuite_etudes"]).strip() != "":
        parts.append(f"Poursuites d’études : {row['poursuite_etudes']}")
    if pd.notna(row["debouches_professionnels"]) and str(row["debouches_professionnels"]).strip() != "":
        parts.append(f"Débouchés professionnels : {row['debouches_professionnels']}")

    # Sélection (0/1)
    if pd.notna(row["formation_selective"]) and int(row["formation_selective"]) == 1:
        parts.append("Sélection : formation sélective")

    if pd.notna(row["epreuves_selection"]) and str(row["epreuves_selection"]).strip() != "":
        parts.append(f"Épreuves de sélection : {row['epreuves_selection']}")

    return "\n\n".join(parts)
```

```python
df["rag_document"] = df.apply(build_rag_doc, axis=1)
```
